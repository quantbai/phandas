# Phandas Alpha Factor Development

Complete example: develop and backtest a quantitative factor using Phandas.

This notebook demonstrates the full workflow:
1. Fetch market data from Binance
2. Design an alpha factor
3. Run backtest and analyze results


```python
%pip install -q phandas
```

```python
from phandas import *
```

```python
panel = fetch_data(
    symbols=['ETH', 'SOL', 'ARB', 'OP', 'POL', 'SUI'],
    start_date='2022-01-01',
    sources=['binance']
)
```

```python

close = panel['close']
volume = panel['volume']
open_price = panel['open']

# 20-day momentum factor
log_returns = log(close / ts_delay(close, 20))
momentum = log_returns.rank()
momentum.name = "alpha"

# Backtest
bt_results = backtest(
    entry_price_factor=open_price,
    strategy_factor=momentum,
    transaction_cost=(0.0003, 0.0003),
)

# Results
bt_results.plot_equity()
```
